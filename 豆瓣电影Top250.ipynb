```python
# 归一化
def byone(x):
    max = np.max(x)
    min = np.min(x)
    x1 = (x-min)/(max-min)
    return x1
# 进行归一化处理
a = dataframe['评分']
dataframe['评分'] = byone(a)
dataframe
```
